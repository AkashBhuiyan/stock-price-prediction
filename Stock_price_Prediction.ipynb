Data Pre Processing

````python


import math
import numpy as np

def Dataset(Data, Date):

  Train_Data = Data['Price'][Data['Date'] < Date].to_numpy()
  Data_Train = []
  Data_Train_X = []
  Data_Train_Y = []
  for i in range(0, len(Train_Data), 5):
    try:
      Data_Train.append(Train_Data[i : i + 5])
    except:
      pass

  if len(Data_Train[-1]) < 5:
    Data_Train.pop(-1)

  Data_Train_X = Data_Train[0 : -1]
  Data_Train_X = np.array(Data_Train_X)
  Data_Train_X = Data_Train_X.reshape((-1, 5, 1))
  Data_Train_Y = Data_Train[1 : len(Data_Train)]
  Data_Train_Y = np.array(Data_Train_Y)
  Data_Train_Y = Data_Train_Y.reshape((-1, 5, 1))


  Test_Data = Data['Price'][Data['Date'] >= Date].to_numpy()
  Data_Test = []
  Data_Test_X = []
  Data_Test_Y = []
  for i in range(0, len(Test_Data), 5):
    try:
      Data_Test.append(Test_Data[i : i + 5])
    except:
      pass

  if len(Data_Test[-1]) < 5:
    Data_Test.pop(-1)

  Data_Test_X = Data_Test[0 : -1]
  Data_Test_X = np.array(Data_Test_X)
  Data_Test_X = Data_Test_X.reshape((-1, 5, 1))
  Data_Test_Y = Data_Test[1 : len(Data_Test)]
  Data_Test_Y = np.array(Data_Test_Y)
  Data_Test_Y = Data_Test_Y.reshape((-1, 5, 1))

  return Data_Train_X, Data_Train_Y, Data_Test_X, Data_Test_Y
````

Split the Data into Training and Test set <br/>
Training Period: 2009-01-01 - 2021-08-13 <br/>

Testing Period:  2021-08-14 - 2023-08-13 <br/>

````python
AKTX_Date = '2021-08-14'
AKTX_Train_X, AKTX_Train_Y, AKTX_Test_X, AKTX_Test_Y = Dataset(AKTX, AKTX_Date)
````

Model

````python
import tensorflow as tf
from sklearn.metrics import mean_squared_error, mean_absolute_error

def Model():
  model = tf.keras.models.Sequential([
                                      tf.keras.layers.LSTM(200, input_shape = (5, 1), activation = tf.nn.leaky_relu, return_sequences = True),
                                      tf.keras.layers.Dense(100, activation = tf.nn.leaky_relu),
                                      tf.keras.layers.Dense(50, activation = tf.nn.leaky_relu),
                                      tf.keras.layers.Dense(5, activation = tf.nn.leaky_relu)
                                      ])
  return model
````

````python
model = Model()
````

````python
tf.keras.utils.plot_model(model, show_shapes=True)
````

````python
model.summary()
````

Custom Learning Rate

````python
def scheduler(epoch):

  if epoch <= 150:
    lrate = (10 ** -5) * (epoch / 150)
  elif epoch <= 400:
    initial_lrate = (10 ** -5)
    k = 0.01
    lrate = initial_lrate * math.exp(-k * (epoch - 150))
  else:
    lrate = (10 ** -6)

  return lrate
````

````python

epochs = [i for i in range(1, 1001, 1)]
lrate = [scheduler(i) for i in range(1, 1001, 1)]
plt.plot(epochs, lrate)
````

````python
callback = tf.keras.callbacks.LearningRateScheduler(scheduler)
````

Model Fitting

````python
AKTX_Model = Model()
BATC_Model = Model()
BXPH_Model = Model()
GRAE_Model = Model()
LFAR_Model = Model()
SNPR_Model = Model()
````

````python
AKTX_Model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'mse', metrics = tf.keras.metrics.RootMeanSquaredError())
BATC_Model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'mse', metrics = tf.keras.metrics.RootMeanSquaredError())
BXPH_Model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'mse', metrics = tf.keras.metrics.RootMeanSquaredError())
GRAE_Model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'mse', metrics = tf.keras.metrics.RootMeanSquaredError())
LFAR_Model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'mse', metrics = tf.keras.metrics.RootMeanSquaredError())
SNPR_Model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'mse', metrics = tf.keras.metrics.RootMeanSquaredError())
````

### **Alhaj Textile Mills Ltd**

````python
AKTX_hist = AKTX_Model.fit(AKTX_Train_X, AKTX_Train_Y, epochs = 1000, validation_data = (AKTX_Test_X, AKTX_Test_Y), callbacks=[callback])
````

````python
history_dict = AKTX_hist.history

loss = history_dict["loss"]
root_mean_squared_error = history_dict["root_mean_squared_error"]
val_loss = history_dict["val_loss"]
val_root_mean_squared_error = history_dict["val_root_mean_squared_error"]

epochs = range(1, len(loss) + 1)
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_figheight(5)
fig.set_figwidth(15)

ax1.plot(epochs, loss, label = 'Training Loss')
ax1.plot(epochs, val_loss, label = 'Validation Loss')
ax1.set(xlabel = "Epochs", ylabel = "Loss")
ax1.legend()

ax2.plot(epochs, root_mean_squared_error, label = "Training Root Mean Squared Error")
ax2.plot(epochs, val_root_mean_squared_error, label = "Validation Root Mean Squared Error")
ax2.set(xlabel = "Epochs", ylabel = "Loss")
ax2.legend()
````

### Predicting the closing stock price of Alhaj Textile Mills Ltd

````python
AKTX_prediction = AKTX_Model.predict(AKTX_Test_X)
````

````python
plt.figure(figsize=(20, 5))
plt.plot(AKTX['Date'][AKTX['Date'] < '2021-08-14'], AKTX['Price'][AKTX['Date'] < '2021-08-14'], label = 'Training')
plt.plot(AKTX['Date'][AKTX['Date'] >= '2021-08-09'], AKTX['Price'][AKTX['Date'] >= '2021-08-09'], label = 'Testing')
plt.plot(AKTX['Date'][AKTX['Date'] >= '2021-08-14'], AKTX_prediction.reshape(-1), label = 'Predictions')
plt.xlabel('Time')
plt.ylabel('Closing Price')
plt.legend(loc = 'best')
````
